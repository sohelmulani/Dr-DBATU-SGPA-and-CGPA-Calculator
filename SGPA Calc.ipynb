```python
import numpy as np
print("*NOTE-->ALL INPUT SHOULD BE IN LOWER-CASE.<--")

grades=[]
credit=[]
gc=[]


n=int(input("enter how many subjects do you have"))


for i in range(0,n):
    g=input("enter grade of sub {}---->".format(i+1))
    
    grades.append(g)
   
    if grades[i]=='ex':
        gc.append(10)
    if grades[i]=='aa':
        gc.append(9.5) 
    if grades[i]=='ab':
        gc.append(9)    
    if grades[i]=='bb':
        gc.append(8.5)
    if grades[i]=='bc':
        gc.append(8)
    if grades[i]=='cc':
        gc.append(7.5)
    if grades[i]=='cd':
        gc.append(7)
    if grades[i]=='dd':
        gc.append(6.5)
    if grades[i]=='de':
        gc.append(6)
    if grades[i]=='ee':
        gc.append(5)
    if grades[i]=='nc':
        gc.append(0)
        print("------->you lost {} subject<-------".format(i+1))
        
        
    c=int(input("enter credit of sub {}----->".format(i+1)))  
    credit.append(c)

acredit=np.array(credit)
agc=np.array(gc)
a=np.sum([acredit*agc])
atc=np.sum(credit)
sgpa=a/atc

print("\n")
print("----------->Your sgpa for this semister is {}<------------".format(sgpa))
print("\n")
choice=input("Do you want to cheak your cgpa????\nEnter yes or no")
if choice=='yes':
    pcgpa=float(input("Enter previous semister's cgpa"))
    cgpa=(pcgpa+sgpa)/2
    print("------>your cgpa is {}".format(cgpa))
if choice=='no':
    print("thanx!!!!!")
```
